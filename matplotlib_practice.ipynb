```python
#Import dependencies. 
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
#Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total faire in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot.
plt.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Create the plot and add a label for the legen.
plt.plot(x_axis, y_axis, label = "Boston", marker="o", color = "blue", linewidth = 2)

#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0,45)

#Create a title.
plt.title("PyBer Fare By Month")
plt.legend()
```

```python
#Object oriented approach.
ax = plt.axes()
ax.plot(x_axis, y_axis, label="Boston", marker = "D", color = "green", linewidth= 2)

ax.legend()

ax.grid()

ax.set_title("PyBer Fare By Month")

ax.set_xlabel("Date")

ax.set_ylabel("Fare ($)")
```

```python
#Making dots larger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Using list comprehension to make the data larger.
plt.scatter(x_axis, y_axis, s = [data*3 for data in y_axis])
```

```python
#Using the object oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[data*5 for data in y_axis], c="skyblue", alpha=0.80, edgecolors="black", linewidth = 2, label = "Boston")

ax.set_title("PyBer Fare By Month")

ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.set_xlim(0,50)

ax.legend()
```

### Pie Charts

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
#Assigning colors, figsize, and explode.

colors = ["slateblue","magenta","lightblue","green","yellowgreen",
         "greenyellow","yellow","orange","gold","indianred",
         "tomato","mistyrose"]

plt.subplots(figsize=(8,8))

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis,
        explode = explode_values,
        colors = colors,
        labels = x_axis,
        autopct="%.1f%%")

plt.show()
```

### Pie chart with object oriented interface

```python
fig, ax = plt.subplots(figsize = (8,8))

colors = ["slateblue","magenta","lightblue","green","yellowgreen",
         "greenyellow","yellow","orange","gold","indianred",
         "tomato","mistyrose"]

explode = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

ax.pie(y_axis, labels = x_axis,
      autopct="%.1f%%",
      explode = explode,
      colors = colors,
      shadow = True,
      startangle = 90,
      counterclock = False)

plt.show()
```
